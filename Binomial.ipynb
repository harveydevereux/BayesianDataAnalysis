```
N = 1000
Experiments = 1000
Means = np.zeros(N-1)
STDs = np.zeros(N-1)
p = 0.9
HeadsToTails = np.zeros((N-1,Experiments))
for i in range(1,N):
  for j in range(Experiments):
    HeadsToTails[i-1,j] = np.random.binomial(i,p)/i
  Means[i-1] = np.mean(HeadsToTails[i-1,:])
  STDs[i-1] = np.std(HeadsToTails[i-1,:])
```

```
HeadsToTails.shape
```

```
plt.plot(range(1,N),HeadsToTails);
```

```
plt.errorbar(range(1,N),Means,STDs);
```

```
plt.loglog(STDs)
x = list(range(1,N))
y = [(np.max(STDs))*x[i]**(-0.5) for i in range(0,N-1)]
plt.loglog(x,y)
plt.title("x^-0.5")
```
